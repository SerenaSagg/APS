# Teoría
## Modulación 
La modulación en amplitud (AM) es un proceso mediante el cual la amplitud de una señal portadora sinusoidal de alta frecuencia se varía en función de una señal moduladora.

La información se encuentra contenida en la envolvente de la señal resultante, la cual replica la forma de la señal moduladora siempre que no haya sobremodulación. En el dominio frecuencial, la señal AM genera dos bandas laterales simétricas alrededor de la frecuencia de la portadora, lo que implica que el ancho de banda total es el doble del ancho de banda de la señal original .

Este método permite trasladar señales de baja frecuencia a bandas de transmisión más altas, facilitando su propagación a través de distintos canales de comunicación.

## Amplificación y relación con la potencia

La amplificación de una señal consiste en aumentar su amplitud mediante la multiplicación por un factor **k** constante. Este proceso no altera la frecuencia ni la forma de la señal, sino únicamente su amplitud. \begin{equation} x_2(t) = k\,x_1(t)\end{equation}

La potencia de una señal está directamente relacionada con el cuadrado de su amplitud. Por lo tanto, al aumentar la amplitud, la potencia también se incrementa de manera proporcional al cuadrado de dicho aumento.

En términos logarítmicos, utilizados frecuentemente en procesamiento de señales, la amplificación suele expresarse en decibeles (dB), lo que permite cuantificar de forma más conveniente las variaciones de amplitud y potencia. \begin{equation} G_{dB} = 20 \log_{10}\left(\frac{A_2}{A_1}\right)\end{equation}

## Convolución de señales
La convolución es una operación fundamental en el análisis de señales y sistemas lineales e invariantes en el tiempo (LTI), que permite determinar la salida de un sistema a partir de una señal de entrada y su respuesta al impulso. En tiempo discreto, se define como:
\begin{equation} y[n] = \sum_{k=-\infty}^{\infty} x[k] \, h[n - k]\end{equation}

lo que implica una suma de productos entre la señal de entrada y versiones desplazadas de la respuesta al impulso. Esta operación describe cómo el sistema modifica la señal de entrada para producir la salida .

Un caso particular importante es la convolución con una delta discreta δ[n]. La delta actúa como elemento identidad, es decir:
\begin{equation} x[n] * \delta[n] = x[n]\end{equation}

y, si está desplazada:
\begin{equation} x[n] * \delta[n - n_0] = x[n - n_0]\end{equation}

Esto significa que convolucionar con una delta simplemente reproduce la señal original (o una versión desplazada), lo que refleja que la delta representa un sistema ideal que no altera la señal.

## Ejercicio 1
Sintesis de señales y aplicaciones de operaciones. Cada caso posee su respectivo período y número de muestras.

```python
import matplotlib.pyplot as plt
import numpy as np

def mi_sen(vmax, dc, ff, ph, nn, fs):
    Ts = 1/fs
    tt = np.arange(nn) * Ts
    xx = dc + vmax * np.sin(2*np.pi*ff*tt + ph)
    return tt, xx

def cuadrada(A, f, nn, fs):
    tt = np.arange(nn) / fs
    xx = A * np.sign(np.sin(2*np.pi*f*tt))
    return tt, xx

def escalon(A, t0, nn, fs):
    tt = np.arange(nn) / fs
    xx = np.zeros(nn)
    xx[tt >= t0] = A
    return tt, xx

def pulso_rect(A, t0, tf, nn, fs):
    tt, xu1 = escalon(A, t0, nn, fs)
    tt, xu2 = escalon(A, tf, nn, fs)
    xc = xu1 - xu2
    return tt, xc
```

**1. Señal sinusoidal de 2 KHz.**

```python
# Parámetros
fs = 2000001
N = 10000

# 1) Señal original de 2 kHz
tt1, x1 = mi_sen(vmax=1, dc=0, ff=2000, ph=0, nn=N, fs=fs)
P1=np.var(x1)
print(f"Potencia: {P1:.2f}")

# Gráfico 1
plt.figure()
plt.plot(tt1, x1, label="f = 2 kHz")
plt.title("Gráfico 1: Señal senoidal")
plt.xlabel("Tiempo [s]")
plt.ylabel("Amplitud [V]")
plt.legend()
plt.grid()
plt.show()
```

**2. Señal senoidal amplificada 3 dB y desfasada en π/2.**

```python
# 2) Misma señal amplificada +3 dB y desfasada en pi/2
fs2 = 2000001
N2 = 10000
tt2, x2 = mi_sen(vmax=np.sqrt(2), dc=0, ff=2000, ph=np.pi/2, nn=N2, fs=fs2)
P2=np.var(x2)
print(f"Potencia: {P2:.2f}")
# Gráfico 2
plt.figure()
plt.plot(tt2, x2, label="Amplificada +3 dB y desfasada")
plt.title("Gráfico 2: Señal amplificada")
plt.xlabel("Tiempo [s]")
plt.ylabel("Amplitud [V]")
plt.legend()
plt.grid()
plt.show()
```

**3. Señal senoidal modulada en amplitud por otra señal sinusoidal de frecuencia de 1000 KHz.**

```python
# 3) Modulación
# Parámetros
fs3 = 10000000
N3 = 100000

#Señal original
tt3, xx1 = mi_sen(vmax=1, dc=0, ff=2000, ph=0, nn=N3, fs=fs3)
#Portadora
tt3, xm = mi_sen(vmax=1, dc=0, ff=1000000, ph=0, nn=N3, fs=fs3)
# Señal modulada
x3 = (1 + xx1) * xm

P3=np.var(x3)
print(f"Potencia: {P3:.2f}")

# Gráfico 3
plt.figure(figsize=(10,10))
plt.subplot(3, 1,1)
plt.title("Gráfico 3: Modulación")
plt.plot(tt3, xx1, label="Senoidal", color="lime")
plt.xlabel("Tiempo [s]")
plt.ylabel("Amplitud [V]")
plt.legend()
plt.grid()

plt.subplot(3, 1,2)
plt.plot(tt3,xm, label="Señal Portadora", color="aqua")
plt.xlim(0, 0.00001) #zoom
plt.xlabel("Tiempo [s]")
plt.ylabel("Amplitud [V]")
plt.legend()
plt.grid()

plt.subplot(3, 1,3)
plt.plot(tt3,x3, label="Señal Modulada", color="fuchsia")
plt.plot(tt3,(1+xx1), 'b--', label="Señal Portadora")
plt.plot(tt3,-(1+xx1), 'b--')
plt.xlim(0, 0.002) #zoom
plt.xlabel("Tiempo [s]")
plt.ylabel("Amplitud [V]")
plt.legend()
plt.grid()
plt.show()
```

**4. Señal senoidal con efecto de saturación al 75% de su amplitud.**

```python
# 4) Señal saturada al 75%
fs4 = 2000001
N4 = 10000
tt4, x4 = mi_sen(vmax=1, dc=0, ff=2000, ph=0, nn=N4, fs=fs4)
x4 = np.clip(x4, -0.75, 0.75)
P4=np.var(x4)
print(f"Potencia: {P4:.2f}")

# Gráfico 4
plt.figure()
plt.plot(tt4, x4, label="Saturada al 75%")
plt.title("Gráfico 4: Señal saturada")
plt.xlabel("Tiempo [s]")
plt.ylabel("Amplitud [V]")
plt.legend()
plt.grid()
plt.show()
```

**5. Una señal cuadrada de 4KHz.**

```python
# 5) Señal cuadrada
fs5 = 2000001
N5 = 10000
tt5,x5=cuadrada(A=1, f=4000, nn=N5, fs=fs5)
P5=np.var(x5)
print(f"Potencia: {P5:.2f}")

# Gráfico 5
plt.figure()
plt.plot(tt5, x5, label="f = 4KHz")
plt.title("Gráfico 5: Señal cuadrada")
plt.xlabel("Tiempo [s]")
plt.ylabel("Amplitud [V]")
plt.legend()
plt.grid()
plt.show()
```

**6. Un pulso rectangular de 10ms.**

```python
# 6) Pulso rectangular
fs6 = 2000001
N6 = 40000
tt6, x6 = pulso_rect(A=1, t0=0.002, tf=0.012, nn=N6, fs=fs6)

# Gráfico 6
plt.figure()
plt.plot(tt6, x6, label="t=10 ms", color="lime")
plt.title("Gráfico 6: Pulso rectangular")
plt.xlabel("Tiempo [s]")
plt.ylabel("Amplitud [V]")
plt.legend()
plt.grid()
plt.show()
```

# Análisis de Resultados
En primer lugar, en el *gráfico 1* se observa una señal senoidal de frecuencia 2 kHz. Como se trata de una señal periódica que existe para todo tiempo, puede clasificarse como una señal de potencia, ya que posee potencia media finita y energía infinita.

En el *gráfico 2* se representa la misma señal, pero amplificada 3 dB y desfasada en π/2. El aumento de 3 dB implica un aumento de la amplitud en un factor raíz cuadrada de 2, por lo que también aumenta su potencia. Al igual que la anterior, sigue siendo una señal de potencia, ya que continúa siendo periódica.

En el *gráfico 3* se observa una modulación en amplitud. La amplitud de la portadora varía en el tiempo siguiendo la forma de la señal moduladora, sin modificar su frecuencia. Como la señal modulada es periódica, también se trata de una señal de potencia.

En el *gráfico 4* se muestra una señal senoidal con saturación al 75%. Esto significa que la señal es recortada al alcanzar los valores máximos y mínimos permitidos, por lo que deja de ser una senoide perfecta y aparece distorsión. Sin embargo, sigue siendo una señal periódica, por lo tanto también es una señal de potencia.

En el *gráfico 5* se representa una señal cuadrada de 4 kHz. Debido a que es periódica y se repite indefinidamente en el tiempo, posee potencia finita y energía infinita, por lo que se clasifica como una señal de potencia.

Por último, en el *gráfico 6* se observa un pulso rectangular de 10 ms. A diferencia de las señales anteriores, esta señal solo existe durante un intervalo finito de tiempo y luego se anula, se clasifica como una señal de energía.

En cada caso se aplicaron diferentes parámetros de frecuencia de muestreo y cantidad de muestras, asegurando que se cumpla el criterio de Nyquist.

En la segunda parte del trabajo se calculó la salida de diferentes sistemas mediante la convolución de las señales de entrada **x[n]** con una respuesta al impulso **h[n]** conformada por deltas desplazadas.
